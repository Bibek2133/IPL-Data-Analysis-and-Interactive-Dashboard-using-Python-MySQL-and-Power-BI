### Finding the top batsmen

```python
batting = df.groupby('striker').agg(
    runs=('runs_off_bat', 'sum'),
    balls=('runs_off_bat', 'count')
)

dismissals = df[df['player_dismissed'].notna()] \
    .groupby('player_dismissed').size()

batting['dismissals'] = dismissals
batting['dismissals'] = batting['dismissals'].fillna(0)

# calculate average and strike rate
batting['average'] = batting['runs'] / batting['dismissals']
batting['strike_rate'] = (batting['runs'] / batting['balls']) * 100

batting = batting.reset_index()

# Apply qualification filter
qualified_batsmen = batting[
    (batting['runs'] >= 300) &
    (batting['balls'] >= 200) &
    (batting['dismissals'] >= 10)
]
```

```python
def classify_batsman(row):
    # priority to total runs
    if row['runs'] >= 1000:
        return 'Good'
    
    elif row['average'] >= 35 and row['strike_rate'] >= 130:
        return 'Good'
    
    elif row['runs'] >= 500 or row['average'] >= 25:
        return 'Consistent'
    
    else:
        return 'Bad'

qualified_batsmen['category'] = qualified_batsmen.apply(classify_batsman, axis=1)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

qualified_batsmen[['runs_norm','avg_norm','sr_norm']] = scaler.fit_transform(
    qualified_batsmen[['runs','average','strike_rate']]
)

qualified_batsmen['performance_score'] = (
    0.5 * qualified_batsmen['runs_norm'] +
    0.3 * qualified_batsmen['avg_norm'] +
    0.2 * qualified_batsmen['sr_norm']
)

top_batsmen = qualified_batsmen.sort_values(
    by='performance_score',
    ascending=False
).head(20)
```

### Finding the top Bowlers

```python
# total runs conceded per ball
df['total_runs'] = df['runs_off_bat'] + df['extras']

# bowling aggregation
bowling = df.groupby('bowler').agg(
    runs_conceded=('total_runs', 'sum'),
    balls=('ball', 'count')
)

# wickets taken
wickets = df[df['player_dismissed'].notna()] \
    .groupby('bowler').size()

bowling['wickets'] = wickets
bowling['wickets'] = bowling['wickets'].fillna(0)

# calculate overs and economy
bowling['overs'] = bowling['balls'] / 6
bowling['economy'] = bowling['runs_conceded'] / bowling['overs']

bowling = bowling.reset_index()
```

```python
qualified_bowlers = bowling[
    (bowling['overs'] >= 400) &
    (bowling['wickets'] >= 100)
]
```

```python
def classify_bowler(row):
    if row['wickets'] >= 150 or row['economy'] <= 7:
        return 'Good'
    elif row['wickets'] >= 100 or row['economy'] <= 8.5:
        return 'Consistent'
    else:
        return 'Bad'

qualified_bowlers['category'] = qualified_bowlers.apply(classify_bowler, axis=1)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

qualified_bowlers[['wickets_norm','eco_norm','overs_norm']] = scaler.fit_transform(
    qualified_bowlers[['wickets','economy','overs']]
)

# invert economy (lower is better)
qualified_bowlers['eco_norm'] = 1 - qualified_bowlers['eco_norm']

qualified_bowlers['performance_score'] = (
    0.5 * qualified_bowlers['wickets_norm'] +
    0.3 * qualified_bowlers['eco_norm'] +
    0.2 * qualified_bowlers['overs_norm']
)
```

```python
top_bowlers = qualified_bowlers.sort_values(
    by='performance_score',
    ascending=False
).head(10)

top_bowlers
```

### Finding top all rounders 

```python
all_rounders = pd.merge(
    batting,
    bowling,
    left_on='striker',
    right_on='bowler',
    how='inner'
)
```

```python
qualified_all_rounders = all_rounders[
    (all_rounders['runs'] >= 500) &
    (all_rounders['wickets'] >= 80)
]
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

qualified_all_rounders[['runs_norm','avg_norm','wkt_norm','eco_norm']] = scaler.fit_transform(
    qualified_all_rounders[['runs','average','wickets','economy']]
)

# invert economy
qualified_all_rounders['eco_norm'] = 1 - qualified_all_rounders['eco_norm']

qualified_all_rounders['allrounder_score'] = (
    0.4 * qualified_all_rounders['runs_norm'] +
    0.2 * qualified_all_rounders['avg_norm'] +
    0.1 * qualified_all_rounders['wkt_norm'] +
    0.3 * qualified_all_rounders['eco_norm']
)
```

```python
def classify_allrounder(row):
    if row['runs'] >= 700 and row['wickets'] >= 80:
        return 'Good'
    elif row['runs'] >= 500 and row['wickets'] >= 50:
        return 'Consistent'
    else:
        return 'Average'

qualified_all_rounders['category'] = qualified_all_rounders.apply(
    classify_allrounder, axis=1
)
```

```python
top_all_rounders = qualified_all_rounders.sort_values(
    by='allrounder_score',
    ascending=False
).head(10)

top_all_rounders[[
    'striker','runs','average','strike_rate',
    'wickets','economy','allrounder_score'
]]
```
